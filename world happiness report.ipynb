```python
# importing the all the libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
import seaborn as sns
matplotlib.rcParams["figure.figsize"] = (20,10)
import warnings
warnings.filterwarnings("ignore")
```

```python
#loading the csv dataset using pandas
df= pd.read_csv(r'C:\Users\pavan\OneDrive\Documents\question papers upsc\book1.csv')
df
```

```python
#changing the column names
df.columns=['Country','Region','Happiness_rank','Happiness_score','Standard_error','Economy','Family','Health',
            'Freedom','Trust','Generosity','Dystopia_residual']
```

```python
#calling the first five rows
df.head()
```

```python
#finding the shape of dataset
df.shape
```

```python
#finding the info of dataset
df.info()
```

```python
#finding is there any null values. there are zero null values
df.isnull().sum()
```

```python
#here we describe all the mean max count and everything
df.describe()
```

```python
# we are seeing the relationship of all features using hist plot
df[['Happiness_score','Standard_error','Economy','Family','Health',
            'Freedom','Trust','Generosity','Dystopia_residual']].hist(figsize=(18,12), bins=50, grid=False);
```

```python
# using the corr() to find the correlation 
df.corr()
```

```python
# using the heatmap we can find the relation between each feature
f,ax = plt.subplots(figsize = (15,15))
sns.heatmap(df.corr(), annot = True, linewidth = 0.5, fmt='.2f', ax=ax)
plt.show()
```

```python
# here we find the relation between happiness_score and economy using plot 
import matplotlib.pyplot as plt
plt.plot(df['Happiness_score'],df['Economy'])
plt.title('Happiness_score vs Economy')
plt.xlabel('Happiness_score')
plt.ylabel('Economy')
plt.show()
```

```python
# here we find the relation between happiness_score and family using plot
plt.plot(df['Happiness_score'],df['Family'])
plt.title('Happiness_score vs Family')
plt.xlabel('Happiness_score')
plt.ylabel('Family')
plt.show()
```

```python
# we are finding the mean of happiness_score of each region
df.groupby('Region')['Happiness_score'].mean()
```

```python
# here we are using boxplot to see happiness_score
plt.figure(figsize=(18,5))
sns.set(style='whitegrid')
sns.boxplot(df['Happiness_score'])
plt.show()
```

```python
# here we are seeing box plot of happiness_score of each region
plt.figure(figsize=(18,5))
sns.boxplot(x=df['Region'],y=df['Happiness_score'])
plt.xticks(rotation=75)
plt.show()
```

```python
# here we are seeing box plot and swarmplot of  happiness_score of each region
plt.figure(figsize=(18,5))
sns.boxplot(x=df['Region'],y=df['Happiness_score'])
plt.xticks(rotation=75)
sns.swarmplot(x=df['Region'],y=df['Happiness_score'],color=".25")
plt.xticks(rotation=75)
plt.show()
```

```python
# here we are seeing swarmplot of each feature
sns.set(style="whitegrid")

sns.swarmplot(y=df["Happiness_score"],color='red')
sns.swarmplot(y=df["Dystopia_residual"],color='blue')
sns.swarmplot(y=df["Economy"],color='yellow')
sns.swarmplot(y=df["Trust"],color='purple')
sns.swarmplot(y=df["Health"],color='green')

plt.title('Happiness_score & Dystopia_residual & Economy  & Trust & Health')
plt.show()
```

```python
#counting for each region
sns.countplot(df['Region'])
plt.xticks(rotation=75)
plt.show()
```

```python
# disttibution plot of economy
ax = sns.distplot(df['Economy'], rug=True, hist=False)
plt.show()
```

```python
# distribution plot of health
ax = sns.distplot(df['Health'], vertical=True)
plt.show()
```

```python
# distribution plot of trust
ax = sns.distplot(df['Trust'])
plt.show()
```

```python
# distribution plot of family
ax = sns.distplot(df['Family'], color="y")
plt.show()
```
